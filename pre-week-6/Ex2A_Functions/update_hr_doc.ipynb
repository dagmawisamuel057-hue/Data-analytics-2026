## Lab 3. Employee data

This lab exercise makes updates to information extracted from an employee database.

```python
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'), 
    ('0121', 'IT', 'Mark Blick', '67000'), 
    ('0068', 'IT', 'Kahna Larsen', '112000'), 
    ('0234', 'OPS', 'Jim Smith', '54000')
]
```

```python
# Update Mark's last name from Blick to Blick-Hawley
hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000')
```

```python
print(hr_list)
```

```python
# Update Jim's department from OPS to CS and salary from 54000 to 60000
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
```

```python
print(hr_list)
```

```python
for employee_num, dept_code, name, salary in hr_list:
    formatted_salary = format(int(salary), ",")
    print(f"{employee_num} | {dept_code} | {name} | {formatted_salary}")
```
