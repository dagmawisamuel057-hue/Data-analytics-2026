# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(part_code):
    colon_position = part_code.find(":")
    supplier = part_code[:colon_position]
    return supplier
```

```python
def get_product_num(part_code):
    colon_position = part_code.find(":")
    dash_position = part_code.find("-")
    product_num = part_code[colon_position + 1:dash_position]
    return product_num
```

```python
def get_size(part_code):
    dash_position = part_code.find("-")
    size = part_code[dash_position + 1:]
    return size
```

```python
part_code1 = "ACME:123-L"
part_code2 = "DI:12345-M"
part_code3 = "ACE:1-12"
```

```python
supplier = get_supplier_code(part_code1)
product_num = get_product_num(part_code1)
size = get_size(part_code1)

print(f"For {part_code1}, the supplier is {supplier}. Part number {product_num} is available in size {size}.")
```

```python
supplier = get_supplier_code(part_code2)
product_num = get_product_num(part_code2)
size = get_size(part_code2)

print(f"For {part_code2}, the supplier is {supplier}. Part number {product_num} is available in size {size}.")
```

```python
supplier = get_supplier_code(part_code3)
product_num = get_product_num(part_code3)
size = get_size(part_code3)

print(f"For {part_code3}, the supplier is {supplier}. Part number {product_num} is available in size {size}.")
```

```python
part_codes = ["ACME:123-L", "DI:12345-M", "ACE:1-12"]

for code in part_codes:
    supplier = get_supplier_code(code)
    product_num = get_product_num(code)
    size = get_size(code)

    print(f"For {code}, the supplier is {supplier}. Part number {product_num} is available in size {size}.")
```
